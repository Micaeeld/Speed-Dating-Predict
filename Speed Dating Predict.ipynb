```python
import pandas as pd
```

```python
df = pd.read_csv('data/speeddating.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop('has_null', axis = 1, inplace = True)
```

```python
df.select_dtypes(include = 'object')
```

```python
from sklearn.preprocessing import OrdinalEncoder

for col in df.columns:
    if df[col].dtype == "object" :
        oe = OrdinalEncoder()
        oe.fit(df[[col]])
        df[col] = oe.fit_transform(df[[col]])
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

```python
X = df.drop(['match'], axis = 1)
y = df['match']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.25, random_state = 0)

dt = DecisionTreeClassifier(max_depth = 5, random_state = 0)

dt.fit(xtrain, ytrain)
dt.score(xtest, ytest)
```

### Obviously it would be very easy to have 100% accuracy with a df that has 2 columns that have the decision of the two people ("decision", "decision_o"), so I will remove these two columns from the training to make things more interesting

```python
X = df.drop(['match', 'decision', 'decision_o'], axis = 1)
y = df['match']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier

xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.25, random_state = 0)

gbc = GradientBoostingClassifier(max_depth = 4, random_state = 0)

gbc.fit(xtrain, ytrain)
pred = gbc.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score

pd.crosstab(ytest, pred, rownames = ['Real'], colnames = ['Predito'], margins = True)
```

### Acurácia

```python
from sklearn.metrics import accuracy_score

acuracia = accuracy_score(ytest, pred)
print('Acurácia: %f' % acuracia)
```

### Precision

```python
from sklearn.metrics import precision_score

precision = precision_score(ytest, pred)
print('Precision: %f' % precision)
```

### Recall

```python
from sklearn.metrics import recall_score

recall = recall_score(ytest, pred)
print('Recall: %f' % recall)
```

### F1-Score

```python
from sklearn.metrics import f1_score

f1 = f1_score(ytest, pred)
print('F1-Score: %f' % f1)
```
